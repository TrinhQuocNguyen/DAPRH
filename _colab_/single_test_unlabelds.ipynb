```python
#@title def plot
from matplotlib import pyplot as plt


def plot_with(path, dic, dis, pla=None, threshold=9999):
    q = path
    matches = dic[q][0].split(";")
    dismat  = dis[q][0].split(";")
    if pla == None: pla = [m.split("/")[-2] if len(m) > 3 else m for m in matches]
    else: pla     = pla[q][0].split(";")
    import cv2
    # showing image
    fig = plt.figure(figsize=(10, 7))
    Image_src = cv2.imread(q)  
    Image_src = cv2.cvtColor(Image_src, cv2.COLOR_BGR2RGB)
    
    plt.imshow(Image_src)
    plt.axis('off')
    plt.title("source image")  
    # create figure
    fig = plt.figure(figsize=(30, 30))
    # setting values to rows and column variables
    rows = 1
    columns = len(matches)
    # reading images
    for i, fn in enumerate(matches[:-1]):
        if float(dismat[i]) > threshold: break
        Image1 = cv2.imread(fn)
        Image1 = cv2.cvtColor(Image1, cv2.COLOR_BGR2RGB)
        ax_i = fig.add_subplot(rows, columns, i+1)
        ax_i.set_title("L{}-D{}".format(pla[i],round(float(dismat[i]), 3)))
        plt.axis('off')
        plt.imshow(Image1)
```
