```python
#!sh scripts/pretrain.sh dukemtmc market1501 resnet50 1
!source /content/P2LR-main/theanoEnv/bin/activate; python examples/source_pretrain.py \
    -ds "dukemtmc|msmt17|lpw|prai|caviara" -dt "market1501" \
    -a "resnet50"  --features 2048 --iters 500 --print-freq 1000\
	  --num-instances 4 -b 64 -j 4 --seed 1 --margin 0.0 \
    --warmup-step 10 --lr 0.00035 --milestones 40 70  --epochs 80 --eval-step 2 \
	  --logs-dir "/content/drive/MyDrive/ducanh/Nautilus/projects/P2LR/phase1-500" \
    --data-dir "/content/P2LR-main/datasets" \
    -pt "" \
    # --resume "/content/drive/MyDrive/ducanh/Nautilus/projects/P2LR/phase1-500/checkpoint.pth.tar"
    
```
